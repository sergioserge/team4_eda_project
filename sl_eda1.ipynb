```python
import pandas as pd

EXCEL_FILE = "data/Muesli Project raw data 21-3.xlsx"

def import_xls(file, sheet, header=0):
    df = pd.read_excel(file, sheet_name = sheet, header=header)
    return df

    
```

```python
df_orders = import_xls(EXCEL_FILE, "Orders", header=1)
df_order_process = import_xls(EXCEL_FILE, "Order Process Data")
df_intern = import_xls(EXCEL_FILE, "InternData Study")
df_campaign = import_xls(EXCEL_FILE, "Campaign Data")
df_list = [df_orders, df_order_process, df_intern, df_campaign]
```

```python
def column_rename(df):
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace(' ', '_')
    return df
```

```python
for df in df_list:
    column_rename(df)
    display(df)
```

```python
df.drop("order_date_y", axis=1, inplace =True)
```

```python
df.rename(columns = {"order_date_x": "order_date"}, inplace=True)
display(df)
```

```python
df["pretransportation_duration"] = df["on_truck_scan_date"] - df["order_date"]
df["pretransportation_duration"] = df["pretransportation_duration"].dt.days
```

```python
df.describe()
```

```python
df['weekday'] = df['order_date'].dt.weekday
```

```python
df.head()
```

```python
# Create two labels of delivery date depending on express/ non/express processing(see flowchart)
same_arrival_day = [0, 2, 5, 6]
different_arrival_day = [1, 3, 4]

# Devide the dataframe into two groups with regard to these labels
same_day = df[df['weekday'].isin(same_arrival_day)]
different_days = df[df['weekday'].isin(different_arrival_day)]

# Doublecheck if the division is done correctly
check_difference = df['order_id'].count() - same_day['order_id'].count() - different_days['order_id'].count()
display(f'The Difference should be zero: ', check_difference)
```

```python
# Compare descriptive statistics of each dataset
display('Same Day' ,same_day.describe())
display('Different Day' ,different_days.describe())

# Comparison:
# quantities: shows very equal distribution, though same_day orders are from 4 days and different days are from 3
 # Tuesday, Thursday and Friday are days with a lot of incoming orders
```

```python
standard_pre = df[df['processing_mode']=='Standard Processing']['pretransportation_duration']
express_pre = df[df['processing_mode']=='Express']['pretransportation_duration']

display('standard_pre ', standard_pre.describe())
display('express_pre ', express_pre.describe())
```

```python
# add column with weekday
df['weekday_name'] = df['weekday'].map({
    0: 'Monday', 
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday' })

df['weekday_name'].value_counts().plot(kind='bar', rot=0, figsize=(8, 6))
```

```python
# Select only orders which would arrive before weekend --> only monday with standard preprocessing
standard_before_weekend = df[df['weekday'] == 0 & df['processing_mode'].isin(['Standard Processing'])]
display('standard_before_weekend', standard_before_weekend['pretransportation_duration'].describe())
display('mode', standard_before_weekend['pretransportation_duration'].mode())

# Select only orders which would arrive before weekend --> only monday and tuesday with express preprocessing
express_before_weekend = df[df['weekday'].isin([0, 1]) & df['processing_mode'].isin(['Express'])]
display('express_before_weekend', express_before_weekend['pretransportation_duration'].describe())
display('mode', express_before_weekend['pretransportation_duration'].mode())
```

```python
# Processing Level

# Select only orders which would be on truck before weekend with standard preprocessing
standard_process_kpi = df[df['weekday'].isin([0, 1, 2]) & df['processing_mode'].isin(['Standard Processing'])]
display('standard_process_kpi', standard_process_kpi['pretransportation_duration'].describe())
display('mode', standard_process_kpi['pretransportation_duration'].mode())

# Select only orders which would be on truck before weekend with express preprocessing
express_process_kpi = df[df['weekday'].isin([0, 1, 2, 3, 4]) & df['processing_mode'].isin(['Express'])]
display('express_process_kpi', express_process_kpi['pretransportation_duration'].describe())
display('mode', express_process_kpi['pretransportation_duration'].mode())

# Average processing duration (standard): mean --> 2
# Average processing duration (express): mean --> 2 
# --> visualization (process issues not relevant, besides delivery)
```

```python
# order came in on weekend, monday and tuesday should take much more, since all have to be processed ond monday and tuesday
for i in range(0, 7):
    print(df[df['weekday'] == i]['pretransportation_duration'].mean())

```

```python
# Warehouse Level

# merging of df and df_intern
df_intern_merged = pd.merge(df, df_intern, how='inner', on='order_id')
df_intern_merged.duplicated(subset='order_id', keep='first')
df_intern_merged.info()

# calculating the duration between order and ready for shipment
df_intern_merged['preprocessing_duration'] = df_intern_merged['ready_to_ship_date'] - df_intern_merged['order_date']
df_intern_merged['preprocessing_duration'] = df_intern_merged['preprocessing_duration'].dt.days

# Select only orders which would be ready to ship with both processing modes
warehouse_kpi = df[df['weekday'].isin([0, 1, 2, 3])]
display('warehouse_kpi', warehouse_kpi['pretransportation_duration'].describe())
display('mode', warehouse_kpi['pretransportation_duration'].mode())
# Average warehouse duration: mean --> 2
# --> visualization (all process issues are relevant)


# Visualization of the timeframe of intern data and total data
# Compare 
```
